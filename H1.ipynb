```python
import pandas as pd

def clean_currency(x):
    if isinstance(x, str):
        return float(x.replace('$', '').replace(',', ''))
    return x

def preprocess_listings(df):
    df['price'] = df['price'].apply(clean_currency)
    bool_cols = ['host_is_superhost', 'host_has_profile_pic', 'host_identity_verified', 'instant_bookable']
    for col in bool_cols:
        if col in df.columns:
            df[col] = df[col].map({'t': 1, 'f': 0}).fillna(0)
            
    df['amenity_count'] = df['amenities'].apply(lambda x: len(str(x).split(',')))
    
    return df

def preprocess_calendar(df):
    df['date'] = pd.to_datetime(df['date'])
    df['available'] = df['available'].map({'t': 1, 'f': 0})
    return df

def engineer_calendar_features(df_cal):

    df_cal = df_cal.sort_values(['listing_id', 'date'])
    
    calendar_features = df_cal.groupby('listing_id').agg(
        occupancy_rate_365=('available', lambda x: 1 - x.mean()), # If available=1, occupied=0
        busy_days_count=('available', lambda x: (x==0).sum())
    ).reset_index()
    
    return calendar_features

listings = preprocess_listings(listings)
calendar = preprocess_calendar(calendar)
calendar_features = engineer_calendar_features(calendar)
listings = listings.merge(calendar_features, left_on='id', right_on='listing_id', how='left')
listings.drop(columns=['listing_id'], inplace=True)
display(listings.head())
```

```python
import plotly.express as px

def generate_scatter_map(df):
    
    fig_map = px.scatter_map(
        df, 
        lat="latitude", lon="longitude", 
        color="price", size="accommodates",
        color_continuous_scale=px.colors.sequential.Plasma, 
        range_color=[0, df['price'].quantile(0.95)],
        size_max=15, zoom=11,
        map_style="carto-positron",
        title="<b>Geographical & Size Pricing:</b> How City location and flat size affect pricing"
    )

    fig_map.show()

generate_scatter_map(listings)
```
